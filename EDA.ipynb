```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Note this is from the reduced dataset, not actual pure stats

```python
df = pd.read_csv('/processed_data.csv')
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(data=df.drop(['sex','citizen','route1','nation'],axis=1).corr(),
            annot=False, linewidths=0.5, ax=ax)
```

```python
sns.histplot(data=df, x='myear', binwidth=1, hue='msuccess', multiple='stack')
```

```python
sns.histplot(data=df.hired_ratio, binwidth=0.2)
# need just exped id this does not show much
```

```python
plt.figure(figsize=(32,4))
sns.countplot(data=df, x=df.citizen,
              order = df['citizen'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
 # too many countries, top is nepal
```

```python
sns.countplot(x=df.msuccess, hue=df.death)
```
